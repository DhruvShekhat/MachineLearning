```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Data_Full_Class.csv")
df
```

```python
x1 = df['ssc_p'].values.reshape(-1,1)
x2 = df['hsc_p'].values.reshape(-1,1)
y = df['salary'].values.reshape(-1,1)
```

```python
%matplotlib inline
plt.figure(figsize = (12,9))
plt.scatter(x1, y, color = 'red')
plt.scatter(x2, y, color = 'blue')
plt.xlabel("ssc and hsc")
plt.ylabel("salary")
```

```python
    t_zero = 0;
    t_one = 0;
    t_two = 0;
    
    learning_rate = 0.000001
    iters = 1000
    
    m = len(x1)
    
    for i in range(iters):
        h_of_theta = t_zero + (t_one * x1) + (t_two * x2)
        
        d_of_zero = (1/m) * sum(h_of_theta - y)
        d_of_one = (1/m) * sum(x1 * (h_of_theta - y))
        d_of_two = (1/m) * sum(x2 * (h_of_theta - y))
        
        t_zero = t_zero - learning_rate * d_of_zero
        t_one = t_one - learning_rate * d_of_one
        t_two = t_two - learning_rate * d_of_two
        
        cost = (0.5 / m) * sum([val**2 for val in(h_of_theta - y)])
        
        print("t_zero: {}, t_one: {}, t_two: {}, cost: {}, iters: {}".format(t_zero, t_one, t_two, cost, i))
```

```python
h = t_zero + (t_one * x1) + (t_two * x2)
%matplotlib inline
plt.figure(figsize = (12,9))
plt.scatter(x1, y, color = 'red')
plt.scatter(x2, y, color = 'blue')
plt.plot(x2, h, color = "green")
plt.xlabel("ssc and hsc")
plt.ylabel("salary")
```
